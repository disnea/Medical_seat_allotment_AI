```python
import pandas as pd
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments

# Load your data from Excel file
data = pd.read_excel(r"C:\Users\Yaswa\Downloads\New folder\MCI_1st_LIST_ALLOTMENT ALLINDIA.xlsx")
```

```python
unnamed_columns = [col for col in data.columns if col.startswith('Unnamed')]
data = data.drop(columns=unnamed_columns)

data
```

```python
import pandas as pd
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments
import torch
from torch.utils.data import Dataset

# Split the data into training and validation sets
train_size = 0.8
train_dataset = data.sample(frac=train_size, random_state=42)
val_dataset = data.drop(train_dataset.index).reset_index(drop=True)

# Tokenize the data
tokenizer = T5Tokenizer.from_pretrained("t5-base")

# Create custom dataset class
class CustomDataset(Dataset):
    def __init__(self, tokenizer, data, max_length):
        self.input_ids = []
        self.attention_masks = []
        for _, row in data.iterrows():
            input_text = " ".join(str(value) for value in row)
            encoding = tokenizer.encode_plus(input_text, max_length=max_length, pad_to_max_length=True, truncation=True, return_tensors="pt")
            self.input_ids.append(encoding["input_ids"])
            self.attention_masks.append(encoding["attention_mask"])
        self.input_ids = torch.cat(self.input_ids, dim=0)
        self.attention_masks = torch.cat(self.attention_masks, dim=0)

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {"input_ids": self.input_ids[idx], "attention_mask": self.attention_masks[idx]}

# Fine-tune the model
model = T5ForConditionalGeneration.from_pretrained("t5-base")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

training_args = TrainingArguments(
    output_dir="./t5-finetuned",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=8,
    save_steps=10_000,
    save_total_limit=2,
)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=CustomDataset(tokenizer, train_dataset, 512),
    eval_dataset=CustomDataset(tokenizer, val_dataset, 512),
)
trainer.train()

# Define the response generation function
def generate_response(input_query):
    input_text = input_query
    input_ids = tokenizer.encode(input_text, return_tensors="pt")
    output_ids = model.generate(input_ids, max_length=50, num_beams=4, early_stopping=True)
    generated_response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return generated_response

# Example usage
query = "Provide information about allotted quota, institute, course, category, and candidate."
response = generate_response(query)
print("Generated Response:", response)
```
